## Clean Trip

```python
clean_trip = pd.read_csv('clean_trip.csv')
```

```python
# 定义网格尺寸
n_rows = 30
n_columns = 20
lon_width = (87.913624596 - 87.534902901) / n_columns
lat_height = (42.021223593 - 41.650221676) / n_rows

# 计算网格编号
def calculate_grid_id(lon, lat):
    lon_idx = int((lon + 87.913624596) / lon_width)
    lat_idx = int((42.021223593 - lat) / lat_height)
    return lon_idx + lat_idx * n_columns

# 为每个行程添加网格编号
clean_trip['pickup_grid_id'] = clean_trip.apply(lambda x: calculate_grid_id(x['Pickup Centroid Longitude'], x['Pickup Centroid Latitude']), axis=1)
clean_trip['dropoff_grid_id'] = clean_trip.apply(lambda x: calculate_grid_id(x['Dropoff Centroid Longitude'], x['Dropoff Centroid Latitude']), axis=1)

# 打印更新后的DataFrame信息
print(clean_trip[['pickup_grid_id', 'dropoff_grid_id']].head())
```

```python
# 按pickup_grid_id分组并计算每组的数据数
grouped_pickup_counts = clean_trip.groupby('pickup_grid_id').size()

# 将结果转换为DataFrame并重置索引，方便查看
grouped_pickup_counts_df = grouped_pickup_counts.reset_index(name='count')

# 打印结果
print(grouped_pickup_counts_df)
```

```python
# 读取映射表CSV文件并将其转换回字典格式

# 读取CSV文件到DataFrame
mapping_df_loaded = pd.read_csv('grid_id_mapping.csv')

# 将DataFrame转换为字典
grid_id_mapping = dict(zip(mapping_df_loaded['Old_Grid_ID'], mapping_df_loaded['New_Grid_ID']))

grid_id_mapping
```

```python
# 为数据集应用新的网格编号
clean_trip['new_pickup_grid_id'] = clean_trip['pickup_grid_id'].map(grid_id_mapping)
clean_trip['new_dropoff_grid_id'] = clean_trip['dropoff_grid_id'].map(grid_id_mapping)

# 删除那些新网格编号为NaN的行
filtered_trip_df = clean_trip.dropna(subset=['new_pickup_grid_id', 'new_dropoff_grid_id'])
```

```python
# 按pickup_grid_id分组并计算每组的数据数
grouped_pickup_counts = filtered_trip_df.groupby('new_pickup_grid_id').size()

# 将结果转换为DataFrame并重置索引，方便查看
grouped_pickup_counts_df = grouped_pickup_counts.reset_index(name='count')

# 打印结果
print(grouped_pickup_counts_df)
```

## Pickup DF

```python
import pandas as pd

# 设置开始和结束日期
start_date = '2023-06-01 00:00:00'
end_date = '2023-12-31 23:59:59'

# 生成时间序列，间隔为30分钟
time_slices = pd.date_range(start=start_date, end=end_date, freq='30T')

time_slices
```

```python
pickup_df = pd.read_csv('pickup_df.csv')
pickup_df.info()
```

```python
# 将Trip Start Timestamp列转换为Datetime类型
pickup_df['Trip Start Timestamp'] = pd.to_datetime(pickup_df['Trip Start Timestamp'])
```

```python
# 创建一个新列表示时间片段（每30分钟）
pickup_df['time_slice'] = pickup_df['Trip Start Timestamp'].dt.floor('30min')

# 按照时间片段和网格编号进行分组，并统计每个分组中的记录数
grouped = pickup_df.groupby(['time_slice', 'new_pickup_grid_id']).agg({'new_pickup_grid_id': 'count'})
```

```python
grouped.head(10)
```

```python
grouped.tail(10)
```

```python
import numpy as np

# 将结果重塑为所需的形状
# time_slices = sorted(pickup_df['time_slice'].unique())  # 获取所有时间片段
grid_ids = list(range(220))  # 获取所有网格编号
pick_up = np.zeros((len(time_slices), len(grid_ids)), dtype=int)  # 创建空的结果数组

# 填充结果数组
for i, time_slice in enumerate(time_slices):
    for j, grid_id in enumerate(grid_ids):
        try:
            count = grouped.loc[(time_slice, grid_id), 'new_pickup_grid_id']
        except KeyError:
            count = 0
        pick_up[i, j] = count

# 输出结果
print(pick_up)
```

```python
np.save('pickup.npy', pick_up.T)
```
