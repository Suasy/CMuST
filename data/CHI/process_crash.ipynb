## Clean Crash

```python
clean_crash = pd.read_csv('clean_crash.csv')
```

```python
clean_crash.info()
```

```python
clean_crash.head
```

```python
import pandas as pd

# 假设这是加载数据的代码
# df = pd.read_csv('path_to_your_data.csv')

# 将时间戳列转换为datetime类型
clean_crash['CRASH_DATE'] = pd.to_datetime(clean_crash['CRASH_DATE'])

# 找出时间范围
start_time_min = clean_crash['CRASH_DATE'].min()
end_time_max = clean_crash['CRASH_DATE'].max()

print(f"时间范围从 {start_time_min} 到 {end_time_max}")
```

```python
# 定义网格尺寸
n_rows = 30
n_columns = 20
lon_width = (87.913624596 - 87.534902901) / n_columns
lat_height = (42.021223593 - 41.650221676) / n_rows

# 计算网格编号
def calculate_grid_id(lon, lat):
    lon_idx = int((lon + 87.913624596) / lon_width)
    lat_idx = int((42.021223593 - lat) / lat_height)
    return lon_idx + lat_idx * n_columns

# 为每个行程添加网格编号
clean_crash['grid_id'] = clean_crash.apply(lambda x: calculate_grid_id(x['LONGITUDE'], x['LATITUDE']), axis=1)

# 打印更新后的DataFrame信息
print(clean_crash[['grid_id']].head())
```

```python
# 按pickup_grid_id分组并计算每组的数据数
grouped_counts = clean_crash.groupby('grid_id').size()

# 将结果转换为DataFrame并重置索引，方便查看
grouped_counts_df = grouped_counts.reset_index(name='count')

# 打印结果
print(grouped_counts_df)
```

```python
# 按照计数排序
sorted_grouped_counts_df = grouped_counts_df.sort_values(by='count', ascending=False)

# 打印排序后的DataFrame以确认
print(sorted_grouped_counts_df)

# 输出到TXT文件
sorted_grouped_counts_df.to_csv('sorted_counts.txt', index=False, sep='\t')
```

```python
# 筛选出大于或等于100的网格编号
valid_grid_ids = grouped_counts_df[grouped_counts_df['count'] >= 100]['grid_id']

# 使用这些网格编号过滤原始数据集
filtered_crash_df = clean_crash[clean_crash['grid_id'].isin(valid_grid_ids)]

# 打印过滤后的DataFrame以确认
print(filtered_crash_df.info())
```

```python
# 创建新的网格编号映射
grid_id_mapping = {old_id: new_id for new_id, old_id in enumerate(valid_grid_ids)}

# 为数据集应用新的网格编号
filtered_crash_df['new_grid_id'] = filtered_crash_df['grid_id'].map(grid_id_mapping)
```

```python
# 将映射转换为DataFrame
mapping_df = pd.DataFrame(list(grid_id_mapping.items()), columns=['Old_Grid_ID', 'New_Grid_ID'])

# 将映射表保存为CSV文件
mapping_file_path = 'grid_id_mapping.csv'
mapping_df.to_csv(mapping_file_path, index=False)
```

```python
# 计算每个网格中心的坐标
grid_centers = []
for grid_id in valid_grid_ids:
    lat_idx = grid_id // n_columns
    lon_idx = grid_id % n_columns
    center_lon = -87.913624596 + lon_idx * lon_width + lon_width / 2
    center_lat = 42.021223593 - lat_idx * lat_height - lat_height / 2
    grid_centers.append((center_lon, center_lat))
```

```python
# 将网格中心坐标转换为DataFrame
grid_centers_df = pd.DataFrame(grid_centers, columns=['Longitude', 'Latitude'])
grid_centers_df['New_Grid_ID'] = grid_centers_df.index  # 添加新网格编号

# 打印网格中心坐标表
print(grid_centers_df)

# 保存网格中心坐标表到CSV
grid_centers_df.to_csv('crash_grid_centers.csv', index=False)
```

```python
import matplotlib.pyplot as plt

# 假设grid_centers_df是包含所有网格中心坐标的DataFrame
# 如果已经保存为CSV，可以使用以下代码重新加载
# grid_centers_df = pd.read_csv('grid_centers.csv')

# 绘制网格中心点
plt.figure(figsize=(3, 8))
plt.scatter(grid_centers_df['Longitude'], grid_centers_df['Latitude'], marker='o', color='b', label='Grid Centers')

# 设置图表标题和标签
plt.title('Grid Centers')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)  # 显示网格线

# 添加图例
plt.legend()

# 显示图形
plt.show()
```

## Filtered CRASH DF

```python
import pandas as pd

filtered_crash_df = pd.read_csv('filtered_crash_df.csv')
```

```python
filtered_crash_df.head
```

```python
# 计算风险评分
filtered_crash_df['risk'] = (filtered_crash_df['INJURIES_FATAL'] * 5 +
                    filtered_crash_df['INJURIES_INCAPACITATING'] * 4 +
                    filtered_crash_df['INJURIES_NON_INCAPACITATING'] * 3 +
                    filtered_crash_df['INJURIES_REPORTED_NOT_EVIDENT'] * 2 +
                    filtered_crash_df['INJURIES_NO_INDICATION'] * 1) + 1
```

```python
filtered_crash_df
```

```python
import pandas as pd

# 设置开始和结束日期
start_date = '2023-06-01 00:00:00'
end_date = '2023-12-31 23:59:59'

# 生成时间序列，间隔为30分钟
time_slices = pd.date_range(start=start_date, end=end_date, freq='30T')

time_slices
```

```python
# 将CRASH_DATE列转换为Datetime类型
filtered_crash_df['CRASH_DATE'] = pd.to_datetime(filtered_crash_df['CRASH_DATE'])
```

```python
# 创建一个新列表示时间片段（每30分钟）
filtered_crash_df['time_slice'] = filtered_crash_df['CRASH_DATE'].dt.floor('30min')

# 按照时间片段和网格编号进行分组，并统计每个分组中的风险数
grouped = filtered_crash_df.groupby(['time_slice', 'new_grid_id']).agg({'risk': 'sum'})
```

```python
grouped.head(10)
```

```python
grouped.tail(10)
```

```python
import numpy as np

# 将结果重塑为所需的形状
# time_slices = sorted(pickup_df['time_slice'].unique())  # 获取所有时间片段
grid_ids = sorted(filtered_crash_df['new_grid_id'].unique())  # 获取所有网格编号
crash_risk = np.zeros((len(time_slices), len(grid_ids)), dtype=int)  # 创建空的结果数组

# 填充结果数组
for i, time_slice in enumerate(time_slices):
    for j, grid_id in enumerate(grid_ids):
        try:
            risk = grouped.loc[(time_slice, grid_id), 'risk']
        except KeyError:
            risk = 0
        crash_risk[i, j] = risk

# 输出结果
print(crash_risk)
```

```python
crash_risk.shape
```

```python
np.save('risk.npy', crash_risk.T)
```
